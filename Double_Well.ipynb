```python
# basic import
%matplotlib inline
from pylab import *
from numpy import *
```

```python
# physical constants (atomic units)
hbar = 1.0
m = 1.0
```

```python
# function to assemble the matrix Hamiltonian
from scipy.sparse import dia_matrix, eye, kron
from numpy import ones

# matrix of 2nd differences
def d2matrix(nelem):
    elements = ones((3,nelem))
    elements[1,:] *= -2
    return dia_matrix((elements, [-1,0,1]), shape=(nelem,nelem)).tocsc()

# discrete Hamiltonian in x representation
def make_H(V):
    nx = len(V)
    kinetic = -hbar**2/(2*m)*d2matrix(nx)
    potential = diag(V)
    H = kinetic + potential
    return H
```

```python
#function to solve the time independent Schrodinger equation
def solve(H):
    eigvals, eigvecs = linalg.eig(H)
    
    I = eigvals.argsort()
    eigvals = eigvals[I]
    eigvecs = array(eigvecs)[:,I]
    return eigvals, eigvecs
```

##The symmetric system

```python
# define the symmetric double well pot. on a discrete  grid
nx = 100

minx = -1.
maxx = 1.

x = linspace(minx, maxx, nx)
perturbation = 0.0

V = zeros(nx)
a1 = -0.5
a2 = 0.5
asize = 0.2
apot = -0.07
V[(x>a1-asize) & (x<a1+asize) ] = apot
V[(x>a2-asize) & (x<a2+asize) ] = apot*(1+perturbation)
```

```python
# solve the problem, print the ionization potential and the bound levels w.r.t. E0
eigvals, eigvecs = solve(make_H(V))
IP = eigvals[0]
print("IP = %f\n" % IP)
for i,E in enumerate(eigvals):
    if E<0:
        print("E%d =\t%g" %(i,E-IP))
```

```python
# plot the wavefunctions and the potential
figure(figsize=(6,6))
for i in arange(6):
    plot(x, eigvecs[:,i]*.05+eigvals[i])
plot(x, V, lw=2)
pass
```

###The asymmetric system

```python
# now do the same for a slightly assymmetric potential:
perturbation = 1e-3
V2 = zeros(nx)
V2[(x>a1-asize) & (x<a1+asize) ] = apot
V2[(x>a2-asize) & (x<a2+asize) ] = apot*(1+perturbation)
```

```python
# solve the problem, print the ionization potential and the bound levels w.r.t. E0
eigvals2, eigvecs2 = solve(make_H(V2))
IP = eigvals2[0]
print("IP = %f\n" % IP)
for i,E in enumerate(eigvals2):
    if E<0:
        print("E%d =\t%g" %(i,E-IP))
```

```python
# plot the wavefunctions and the potential
figure(figsize=(6,6))
for i in arange(6):
    plot(x, eigvecs2[:,i]*.05+eigvals2[i])
plot(x, V, lw=2)
pass
```

The ground state eigenfunctions are now well localized.

###Time evolution
For fun, we can also calculate the time evolution of a localized electron wave function $(\psi_u + \psi_g)/\sqrt{2}$ in the symmetric case:
$$
\psi(t) = \sum_j c_j\psi_j(t) = \sum_j c_j e^{{E_j t}/{i \hbar}}\psi_j(0)
$$

```python
coeffs = zeros(nx)
coeffs[0] = 1/sqrt(2.)
coeffs[1] = 1/sqrt(2.)
E_mean = dot(coeffs*eigvals, coeffs)

def psi_t(eigvecs, eigvals, coeffs, t):
    phase = exp(eigvals*t/1j*hbar)
    #print phase
    
    psi = zeros(len(coeffs), dtype=complex)
    for i in arange(len(coeffs)):
        if coeffs[i] != 0.:
            psi += eigvecs[:,i]*phase[i]
    return psi
```

```python
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
```

```python
fig = figure()
ax = axes(xlim=(minx,maxx))
line1, = ax.plot([],[],lw=1)
ax.plot(x, V)


def animate(data):
    line1.set_data(x, abs(psi_t(eigvecs, eigvals, coeffs, 2000000.*data))*.1+E_mean)

    return line1,

anim = animation.FuncAnimation(fig, animate, frames=arange(20), interval=100)
display_animation(anim, default_mode='loop')
```
