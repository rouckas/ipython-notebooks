### Application of the Rapp & Francis theory to D$^-$ + H
http://scitation.aip.org/content/aip/journal/jcp/37/11/10.1063/1.1733066

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import hbar, k, physical_constants
np.seterr(under='ignore')
kB = k
a0 = physical_constants["Bohr radius"][0]
amu = physical_constants["atomic mass constant"][0]
eV = physical_constants["electron volt"][0]
cm2eV = 1.239841930e-4 # http://physics.nist.gov/cgi-bin/cuu/Value?minvev
```

```python
#atomic properties
EA_H = cm2eV*6082.99*eV   # 10.1103/PhysRevA.43.6104
EA_D = cm2eV*6086.6*eV    # 10.1103/PhysRevA.43.6104
IP_He = 24.5874*eV # eV
mu_HD = 2.*1./(2.+1.)*amu
mu_HH = 1.*1./(1.+1.)*amu
mu_He = 2*amu
mu = mu_He
```

```python
I = 0.5*(EA_H + EA_D)
I = IP_He
gamma = np.sqrt(I/(13.6*eV))



omega_H = EA_H/hbar
omega_D = EA_D/hbar
# omega = omega_B - omega_A (A+ + B -> B+ + A)
omega = omega_H - omega_D #  H + D- -> D + H-
```

```python
def sech(x):
    # sech without overflows
    return 2*np.exp(-np.abs(x))/(1+np.exp(-2*np.abs(x)))

class CT:
    """ A+ + B -> B+ + A """
    def __init__(self, I_A, I_B):
        self.I = 0.5*(I_A + I_B)
        self.gamma = np.sqrt(self.I/(13.6*eV))
        self.omega = (I_B - I_A)/hbar
        self.b1 = np.vectorize(self._b1)
        self.sigma_solve = np.vectorize(self._sigma_solve)
        self.f = 1. # statistical factor, not needed for isotopic asymmetry
        
    def A(self, b, v):
        return np.sqrt(2*pi/(self.gamma*a0)) * (self.I/(hbar*v))\
            * b**1.5 * (1+a0/(self.gamma*b)) * np.exp(-self.gamma*b/a0)

    def P(self, b, v):
        return np.sin(self.A(b,v))**2
    
    def perturbation(self, b, v):
        # in eq (51)
        return sech(self.omega/v * np.sqrt(a0*pi*b/(2*self.gamma)))**2
    
    def Pomega(self, b, v):
        return self.f*self.P(b, v)*self.perturbation(b,v)
    
    def p(self, v):
        # adiabatic parameter (fig6)
        return self.omega*a0/v

    def _b1(self, v):
        from scipy.optimize import brentq
        
        if self.omega == 0:
            f = lambda b: self.A(b,v) - pi/6.
            return brentq(f, a0/self.gamma, 1e-5)
        else:
            # find the last maximum of P0:
            f = lambda b: self.A(b,v) - pi/4.
            bmin = brentq(f, a0/self.gamma, 1e-5)
            #print(bmin)

            # solve eq (52) for b>bmin
            f = lambda b: self.perturbation(b,v) - 4*self.P(b,v)
            bmax = 10*bmin
            while f(bmax) <= 0:
                bmax *= 10
                if bmax > 1e-5 : return np.nan
            return brentq(f, bmin, bmax)

    def sigma_approx(self, v, b1_avg):
        b1_approx = -a0/gamma*np.log(v) + a0/(2*gamma)*\
            np.log(72*b1_avg**3/(pi*gamma*a0) * I**2/hbar**2 * (1+a0/(gamma*b1_avg))**2)
        return pi/2*b1_approx**2
    
    def _sigma_solve(self, v):
        
        if self.omega == 0:
            return pi/2*self.b1(v)**2
        else:
            from scipy.integrate import quad
            f = lambda b: self.perturbation(b, v)*2*pi*b
            bmax = self.b1(v)
            if np.isnan(bmax):
                return np.nan
            return 0.5*self.f*quad(f, 0, bmax, epsrel=1e-3)[0]
    
    def sigma_integrate(self, v):
        from scipy.integrate import quad
        f = lambda b: self.P(b, v)*b
        integral = quad(f, 1e-12, np.inf)
        print(integral)
        return integral[0]*2*pi
```

```python
# find the extremum of the phase shift w.r.t. b.
# This is used as lower limit for b1 finding in CT._b1
import sympy as sp
sp.init_printing()
sa0, sb, sgamma = sp.symbols("a0, b, gamma", positive=True)
A = sb**(sp.S(3)/2) * (1+sa0/(sgamma*sb)) * sp.exp(-sgamma*sb/sa0)
Aprime = A.diff(sb)
sp.solve(Aprime, sb)
```

```python
# for illustration plot the phase shift and the oscillations in the intermediate velocity range

dE = IP_He*0.01 # 1% energy perturbation
CT_He = CT(IP_He + dE/2, IP_He - dE/2)

# plot the CT probability at thermal velocity at 100 K
b = np.linspace (1e-12, .2e-7, 20000)
b = np.logspace (-12, -7, 2000)

v = 5e4

subplot(211)
plt.plot(b, CT_He.A(b, v))
plt.gca().set_xscale("log")

subplot(212)
plt.plot(b, CT_He.perturbation(b, v))
plt.plot(b, CT_He.Pomega(b, v))
plt.gca().set_xscale("log")
```

```python
from maxwell import mxw_v, mxw_v_avg
@vectorize
def rate_integrate(T, mu, sigma):
    from scipy.integrate import quad
    f = lambda v: mxw_v(v, T=T, M=mu)*v*sigma(v)
    v_avg = mxw_v_avg(T=T, M=mu)
    return quad(f, v_avg*1e-2, v_avg*1e2)[0]
```

```python
# calculate the rate coeffs
T = np.logspace(0, 3, 100)
rate_HD = rate_integrate(T, mu=mu_HD, sigma = CT_HD.sigma_solve)
rate_HH = rate_integrate(T, mu=mu_HD, sigma = CT_HD0.sigma_solve)
```
