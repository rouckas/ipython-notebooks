```python
from diatoms import H2, HD, D2, eV2cm, cm2J, kB
```

```python
from numpy import *
from matplotlib.pyplot import *
%matplotlib inline
rcParams["mathtext.fontset"] = "stixsans"
```

```python
# equilibrium concentrations of para/ortho D2
T = linspace(1, 200)
plot(T, H2.Zo(T)/H2.Z(T), label="ortho H2")
plot(T, H2.Zp(T)/H2.Z(T), label="para H2")
plot(T, D2.Zo(T)/D2.Z(T), "--", label="ortho D2")
plot(T, D2.Zp(T)/D2.Z(T), "--", label="para D2")
xlabel(r"$T\ (\rm K)$")
ylabel("$f$")
legend(loc="center left")
savefig("H2_D2_populations.pdf")
```

```python
# equilibrium concentrations of para/ortho D2
T = logspace(.5, 3)
H2.set_temp(300)
plot(T, H2.Erot(T), "-", label="normal H$_2$")
H2.f_o = 0.005
plot(T, H2.Erot(T), "-", label="99.5% para-H$_2$")

HD.Jmax=40
plot(T, HD.Erot(T), "--", label="HD")

plot(T, kB*T/cm2J, ":", label=r"$K_{\rm B} T$ - equipartition theorem")
xscale("log")
yscale("log")
ylim(ymin=1e-7)
xlim(xmin=5)
xlabel(r"$T\ (\rm K)$")
ylabel(r"$\left< E_{\rm rot}\right>\ (\rm cm^{-1})$")
legend(loc="lower right")
savefig("H2_HD_levels.pdf")
```
